```python
def get_input_patterns(input: str):
    patterns = []
    current_pattern = []
    for line in input.split("\n"):
        if len(line) > 1:
            current_pattern.append(np.array([char for char in line]))
        else:
            patterns.append(np.array(current_pattern) == "#")
            current_pattern = []
    patterns.append(np.array(current_pattern) == "#")
    return patterns

patterns = get_input_patterns(current_input)
len(patterns)
```

```python
a = patterns[0]
a.shape[0]

def get_rows(a):
    return a.dot(1 << np.arange(a.shape[-1] - 1, -1, -1))

def get_cols(a):
    return a.transpose().dot(1 << np.arange(a.shape[0] - 1, -1, -1))

def get_symmetry(numbers) -> int:
    l = len(numbers)
    for i in range(1, l):
        lh, rh = np.flip(numbers[0:i]), numbers[i:]
        max_len = min(len(lh), len(rh))
        if all(lh[:max_len] == rh[:max_len]):
            return i

    return 0

print(get_symmetry(get_cols(a)))
```

```python
assert 0 == get_symmetry(get_rows(a)) # no sym axis that spans the whole patter:  narray([358,  90, 385, 385,  90, 102, 346])
assert 5 == get_symmetry(get_cols(a)) # sym axis after element 4:  array([ 89,  24, 103,  66,  37,  37,  66, 103,  24])

get_rows(a), get_cols(a)
```

```python
def print_result(patterns):
    num_cols, num_rows = 0, 0
    for pattern in patterns:
        num_cols += get_symmetry(get_cols(pattern))
        num_rows += get_symmetry(get_rows(pattern))
    print("result: ", num_cols + 100 * num_rows)

print_result(patterns)
```

```python
current_input = input
patterns = get_input_patterns(current_input)

import itertools
import math
this_map = patterns[0]

sums = [0, 0]

for this_map in patterns:
    rows: np.array = get_rows(this_map)
    cols = get_cols(this_map)

    orig_map = np.copy(this_map)

    # print(rows)
    # print(replacement := [(a, b, int(math.log2(a ^ b))) for (a, b) in itertools.combinations(rows, 2) if (2**20 + a - b).bit_count() == 2])
    replacement_rows = [(a, b, int(math.log2(a ^ b))) for (a, b) in itertools.combinations(rows, 2) if (2**24 + a - b).bit_count() == 2]
    counter = 0
    if len(replacement_rows) > 0:
        counter += 1
        loc = np.argwhere(rows == replacement_rows[0][0]).flatten()
        # print(this_map[loc[0]])
        # print(this_map, "\n\n")
        prev_symmetris = set()
        prev_symmetris.add((0, get_symmetry(get_rows(this_map))))
        prev_symmetris.add((1, get_symmetry(get_cols(this_map))))
        # print(f"{prev_symmetris = }")

        this_map[loc[0]][-replacement_rows[0][2] - 1] ^= 1
        # print(this_map)

        symmetries = set()
        symmetries.add((0, get_symmetry(get_rows(this_map))))
        symmetries.add((1, get_symmetry(get_cols(this_map))))
        # print(f"{symmetries = }")

        resulting_symmetries = symmetries - prev_symmetris
        if len(resulting_symmetries) > 0:
            # print(resulting_symmetries)
            new_sym = next(iter(resulting_symmetries))
            sums[new_sym[0]] += new_sym[1]

    if len(replacement_cols := [(a, b, int(math.log2(a ^ b))) for (a, b) in itertools.combinations(cols, 2) if (2**24 + a - b).bit_count() == 2]) > 0:
    # if len(replacement_cols) > 0:
        if counter == 1:
            print("second match")
        loc = np.argwhere(cols == replacement_cols[0][0]).flatten()
        # print(orig_map[-replacement_cols[0][2] - 1][loc[0]])
        # print(orig_map, "\n\n")
        # print(cols)
        prev_symmetris = set()
        prev_symmetris.add((0, get_symmetry(get_rows(orig_map))))
        prev_symmetris.add((1, get_symmetry(get_cols(orig_map))))
        # print(f"{prev_symmetris = }")

        orig_map[-replacement_cols[0][2] - 1][loc[0]] ^= 1
        # print(orig_map)

        symmetries = set()
        symmetries.add((0, get_symmetry(get_rows(orig_map))))
        symmetries.add((1, get_symmetry(get_cols(orig_map))))
        # print(f"{symmetries = }")

        resulting_symmetries = symmetries - prev_symmetris
        if len(resulting_symmetries) > 0:
            print(f"{resulting_symmetries=}")
            new_sym = next(iter(resulting_symmetries))
            sums[new_sym[0]] += new_sym[1]

print(sums)
```
